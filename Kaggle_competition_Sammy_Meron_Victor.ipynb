```python
# Checking for outliers

for variables in train_X:
    plt.figure(figsize=(10,4))
    sns.boxplot(data=train_X, x=train_X[variables],
                notch=True, showcaps=False,
                flierprops={"marker": "o"},
                medianprops={"color": "firebrick"})
    plt.show()
```

>We can see that there are a lot of outliers and we decided to delete them all in the begining, but we are going to check if we are going to delete too much data

```python
# Creating a new variable to check

train_X_new = train_X.copy()
test_X_new = test_X.copy()
validation_X_new = validation_X.copy()
```

```python
# Creating a function that gets the outliers and transform them to nan

def outliers_remover(df):
    print("Old Shape: ", df.shape)
    continuous_variables = df.select_dtypes(include='number') # Getting only numerical features
    for column in continuous_variables.columns:
        q75,q25 = np.percentile(df.loc[:,column],[75,25]) # Calculate the interquartiles
        intr_qr = q75-q25 # Interquartile range
        max = q75+(1.5*intr_qr) # Setting the upper edge
        min = q25-(1.5*intr_qr) # Setting the lower edge
        df.loc[df[column] < min,column] = np.nan # Transforming outliers into nans
        df.loc[df[column] > max,column] = np.nan
```

```python
outliers_remover(train_X_new)
outliers_remover(test_X_new)
outliers_remover(validation_X_new)
```

```python
# Removing the nans

train_X_new = train_X_new.dropna().reset_index(drop=True)
test_X_new = test_X_new.dropna().reset_index(drop=True)
validation_X_new = validation_X_new.dropna().reset_index(drop=True)
```

```python
# Checking the new shapes

print("New Shape: ", train_X_new.shape)
print("New Shape: ", test_X_new.shape)
print("New Shape: ", validation_X_new.shape)
```

>We can see here that we are deleting more than 90% of data so we decided that we are going to keep the original data because we don't have as much time to decide which features delete the nan and which ones replacing them for the mean or median

```python
# import XGBoost
import xgboost as xgb


# define data_dmatrix
data_dmatrix_train_X = xgb.DMatrix(data=train_X,label=train_y)
data_dmatrix_test_X = xgb.DMatrix(data=test_X,label=test_y)
data_dmatrix_validation_X = xgb.DMatrix(data=validation_X,label=validation_y)
```

```python
# Renaming variables for our inner peace 

X_train = train_X
y_train = train_y
X_test = test_X
y_test = test_y
X_validation = validation_X
y_validation = validation_y
```

```python
# import XGBClassifier
from xgboost import XGBClassifier


# declare parameters
params = {
            'objective':'binary:logistic',
            'max_depth': 4,
            'alpha': 10,
            'learning_rate': 1.0,
            'n_estimators':100
        }         
           
          
# instantiate the classifier 
xgb_clf = XGBClassifier(**params)


# fit the classifier to the training data
xgb_clf.fit(X_train, y_train)
```

```python
# make predictions on test data

y_pred = xgb_clf.predict(X_test)
```

```python
# compute and print accuracy score

from sklearn.metrics import accuracy_score

print('XGBoost model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))
```

```python
# k-fold Cross Validation using XGBoost

from xgboost import cv

params = {"objective":"binary:logistic",'colsample_bytree': 0.3,'learning_rate': 0.1,
                'max_depth': 5, 'alpha': 10}

xgb_cv_train = cv(dtrain=data_dmatrix_train_X, params=params, nfold=3,
                    num_boost_round=50, early_stopping_rounds=10, metrics="auc", as_pandas=True, seed=123)

xgb_cv_validation = cv(dtrain=data_dmatrix_validation_X, params=params, nfold=3,
                    num_boost_round=50, early_stopping_rounds=10, metrics="auc", as_pandas=True, seed=123)
```
